```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the main data file
data_file_path = 'API_SP.POP.TOTL_DS2_en_csv_v2_144171.csv'
data = pd.read_csv(data_file_path, skiprows=4)
```

```python
# Select the data for the year 2020
year = '2020'
population_2020 = data[['Country Name', year]].dropna()
```

```python
population_2020
```

```python
population_2020.shape
```

```python
population_2020.isnull().sum()
```

```python
# Histogram for Population Distribution in 2020
plt.figure(figsize=(12, 6))
sns.histplot(population_2020[year], bins=30, kde=True)
plt.title('Distribution of Population in 2020')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.show()
```

```python
# Bar Chart for Top 10 Countries by Population in 2020
top_10_countries = population_2020.nlargest(10, year)
plt.figure(figsize=(12, 6))
sns.barplot(x=year, y='Country Name', data=top_10_countries)
plt.title('Top 10 Countries by Population in 2020')
plt.xlabel('Population')
plt.ylabel('Country')
plt.show()
```
